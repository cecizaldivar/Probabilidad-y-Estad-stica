# Crear un índice 

```python
# Vamos a crear una definción para clasificar (buena, mala o regula) canciones de acuerdo a la popularidad 

def rate(Popularity):
  if Popularity >= 80:
    return "Buena"
  elif Popularity >= 70:
    return "Regular"
  else: 
    return "Mala"
```

```python
# Crearemos una columna que se llamará Class 

top50_data2["Class"] = top50_data2["Popularity"].apply(rate)
top50_data2[['Artist.Name', 'Track.Name', 'Genre', 'Length.', 'Popularity', 'Class']]
```
